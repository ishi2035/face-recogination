```python
from sklearn.datasets import fetch_olivetti_faces
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import LeaveOneOut
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn import metrics
import numpy as np
```

```python
oli_data = fetch_olivetti_faces()
```

```python
oli_data
```

```python
features = oli_data.data
tragets = oli_data.target
```

since the value is already minimized no need of scaling it down

```python
fig, sub_plots = plt.subplots(nrows=5, ncols=8, figsize=(14,8))
sub_plots = sub_plots.flatten()
```

```python
for(unique_user_id) in np.unique(tragets):
  image_index = unique_user_id*8
  sub_plots[ unique_user_id].imshow(features[image_index].reshape(64,64), cmap='gray')
  sub_plots[ unique_user_id].set_xticks([])
  sub_plots[ unique_user_id].set_yticks([])
  sub_plots[ unique_user_id].set_title("face id: %s" % unique_user_id)
```

```python
plt.suptitle("THE DataSet of 40 People")
plt.show()
```

x_train = traning data & 
x_test = data for testing&
y = target for both traning and testing

```python
x_train, x_test, y_train, y_test = train_test_split(features, tragets, test_size = 0.25, stratify=tragets,random_state=0) 
```

```python
pca = PCA(n_components=100, whiten=True )
pca.fit(x_train)

x_train_pca= pca.transform(x_train)
x_test_pca=  pca.transform(x_test)
```

```python
 models = [("logistic Regression", LogisticRegression()),("support vector", SVC()), ("naive bayes", GaussianNB())] 
```

```python
for name,model in models:
   model.fit(x_train_pca, y_train)
   y_pred = model.predict(x_test_pca)
   print("results with %s" % name)
   print("accuracy %s" %(metrics.accuracy_score(y_test, y_pred)))
```
